```python
import pandas as pd
import xarray as xr 
import matplotlib.pyplot as plt
import numpy as np

ground_truth = pd.read_csv('cmems_vertical_profiles.csv', header=None).to_numpy()
predicted = pd.read_csv('random_forest_vertical_profs_sss_sst.csv', header=None).to_numpy()
predicted2 = pd.read_csv('random_forest_vertical_profs.csv', header=None).to_numpy()
predicted3 = pd.read_csv('LSTM_35_prof.csv', header=None).to_numpy()
predicted4 = pd.read_csv('LSTM_512_prof.csv', header=None).to_numpy()
predicted6 = pd.read_csv('model-3_prof.csv', header=None).to_numpy()

heights = xr.open_dataset('../val_ds/cmems_mod_glo_phy-all_my_0.25deg_P1D-m-20220202.nc')
heights = heights.depth

'''
def plot_vertical_profile(ground_truth, predicted, heights, outdir, lat, lon, date, varname):
    plt.figure(figsize=((5,10)), dpi=300)
    plt.title(f'{varname} at [{lon}ยบ, {lat}ยบ] for day {date}')
    plt.plot(ground_truth, -heights, label='OGCM')
    plt.plot(predicted, -heights, label='Predicted')
    plt.xlabel(f'{varname}')
    plt.ylabel('Depth (m)')
    plt.legend()
    plt.grid(True)
    plt.savefig(f'{outdir}/{varname[0:4]}_{date}_{lon}_{lat}.png', bbox_inches="tight")
'''

heights = heights[heights < 1000]
np.savetxt("heights.csv", heights, delimiter=",")

```

```python
lat = -0.948
lon = -128.586
date = '2022-10-29'
outdir = '.'

varname = 'ASAL (g/kg)'
plt.figure(figsize=((5,10)), dpi=300)
plt.title(f'{varname} at [{lon}ยบ, {lat}ยบ] for day {date}')
plt.plot(ground_truth[0,0:46], -heights, label='OGCM', color='black', zorder=5, linewidth=3)
plt.plot(predicted[0,0:46], -heights, label='RFRv1', alpha = 0.7)
plt.plot(predicted2[0,0:46], -heights, label='RFRv2', alpha = 0.7)
plt.plot(predicted3[2,0:46], -heights, label='LSTMv1', alpha = 0.7)
plt.plot(predicted6[2,0:46], -heights, label='LSTMv2', alpha = 0.7)
plt.xlabel(f'{varname}')
plt.ylabel('Depth (m)')
plt.legend()
plt.grid(True)
plt.savefig(f'{outdir}/{varname[0:4]}_{date}_{lon}_{lat}.png', bbox_inches="tight")

varname='CTEMP (Cยบ)'
plt.figure(figsize=((5,10)), dpi=300)
plt.title(f'{varname} at [{lon}ยบ, {lat}ยบ] for day {date}')
plt.plot(ground_truth[0,46:], -heights, label='OGCM', color='black', zorder=5, linewidth=3)
plt.plot(predicted[0,46:], -heights, label='RFRv1', alpha = 0.7)
plt.plot(predicted2[0,46:], -heights, label='RFRv2', alpha = 0.7)
plt.plot(predicted3[2,46:], -heights, label='LSTMv1', alpha = 0.7)
plt.plot(predicted4[2,46:], -heights, label='LSTMv2', alpha = 0.7)

plt.xlabel(f'{varname}')
plt.ylabel('Depth (m)')
plt.legend()
plt.grid(True)
plt.savefig(f'{outdir}/{varname[0:4]}_{date}_{lon}_{lat}.png', bbox_inches="tight")
```
