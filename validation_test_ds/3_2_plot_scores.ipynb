```python
import pandas as pd
import matplotlib.pyplot as plt

df =pd.read_csv('random_forest_scores.csv')
y_values = pd.read_csv('heights.csv', header=None)
df2 = pd.read_csv('metrics_32_32.csv')
df3 = pd.read_csv('metrics_512_new.csv')
df4 = pd.read_csv('metrics_model-3.csv')

r2_as = (
    df[['r2_m11_as', 'r2_m12_as']]
    .join(df2[['ASAL R2']].add_suffix('_df2'))
    .join(df4[['ASAL R2']].add_suffix('_df3'))
)

r2_ct = (
    df[['r2_m11_ct', 'r2_m12_ct']]
    .join(df2[['CTEMP R2']].add_suffix('_df2'))
    .join(df3[['CTEMP R2']].add_suffix('_df3'))
)

mse_as = (
    df[['mse_m11_as', 'mse_m12_as']]
    .join(df2[['ASAL MSE']].add_suffix('_df2'))
    .join(df4[['ASAL MSE']].add_suffix('_df3'))
)

mse_ct = (
    df[['mse_m11_ct', 'mse_m12_ct']]
    .join(df2[['CTEMP MSE']].add_suffix('_df2'))
    .join(df3[['CTEMP MSE']].add_suffix('_df3'))
)

mae_as = (
    df[['mae_m11_as', 'mae_m12_as']]
    .join(df2[['ASAL MAE']].add_suffix('_df2'))
    .join(df4[['ASAL MAE']].add_suffix('_df3'))
)

mae_ct = (
    df[['mae_m11_ct', 'mae_m12_ct']]
    .join(df2[['CTEMP MAE']].add_suffix('_df2'))
    .join(df3[['CTEMP MAE']].add_suffix('_df3'))
)


plt.figure(figsize=(4, 5))
plt.plot(r2_as, y_values)
plt.gca().invert_yaxis()
plt.title('R² Score Salinity', fontsize=16)
plt.ylabel('Depth (meters)', fontsize=12)
plt.xlabel('R² Score', fontsize=12)
plt.legend(['RFv1', 'RFv2', 'LSTMv1', 'LSTMv2'], fontsize=10)  # Add legend with model names
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid lines
plt.tick_params(axis='both', which='major', labelsize=10)  # Adjust tick label size
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.savefig('r2_asal.png')
plt.show()

plt.figure(figsize=(4, 5))
plt.plot(r2_ct, y_values)
plt.gca().invert_yaxis()
plt.title('R² Score Temperature', fontsize=16)
plt.ylabel('Depth (meters)', fontsize=12)
plt.xlabel('R² Score', fontsize=12)
plt.legend(['RFv1', 'RFv2', 'LSTMv1', 'LSTMv2'], fontsize=10)  # Add legend with model names
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid lines
plt.tick_params(axis='both', which='major', labelsize=10)  # Adjust tick label size
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.savefig('r2_ctemp.png')
plt.show()

plt.figure(figsize=(4, 5))
plt.plot(mse_as, y_values)
plt.gca().invert_yaxis()
plt.title('MSE for Salinity', fontsize=16)
plt.ylabel('Depth (meters)', fontsize=12)
plt.xlabel('MSE', fontsize=12)
plt.legend(['RFv1', 'RFv2', 'LSTMv1', 'LSTMv2'], fontsize=10)  # Add legend with model names
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid lines
plt.tick_params(axis='both', which='major', labelsize=10)  # Adjust tick label size
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.savefig('mse_asal.png')
plt.show()

plt.figure(figsize=(4, 5))
plt.plot(mse_ct, y_values)
plt.gca().invert_yaxis()
plt.title('MSE for Temperature', fontsize=16)
plt.ylabel('Depth (meters)', fontsize=12)
plt.xlabel('MSE', fontsize=12)
plt.legend(['RFv1', 'RFv2', 'LSTMv1', 'LSTMv2'], fontsize=10)  # Add legend with model names
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid lines
plt.tick_params(axis='both', which='major', labelsize=10)  # Adjust tick label size
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.savefig('mse_ctemp.png')
plt.show()

plt.figure(figsize=(4, 5))
plt.plot(mae_as, y_values)
plt.gca().invert_yaxis()
plt.title('MAE for Salinity', fontsize=16)
plt.ylabel('Depth (meters)', fontsize=12)
plt.xlabel('MAE', fontsize=12)
plt.legend(['RFv1', 'RFv2', 'LSTMv1', 'LSTMv2'], fontsize=10)  # Add legend with model names
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid lines
plt.tick_params(axis='both', which='major', labelsize=10)  # Adjust tick label size
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.savefig('mae_asal.png')
plt.show()

plt.figure(figsize=(4, 5))
plt.plot(mae_ct, y_values)
plt.gca().invert_yaxis()
plt.title('MAE for Temperature', fontsize=16)
plt.ylabel('Depth (meters)', fontsize=12)
plt.xlabel('MAE', fontsize=12)
plt.legend(['RFv1', 'RFv2', 'LSTMv1', 'LSTMv2'], fontsize=10)  # Add legend with model names
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid lines
plt.tick_params(axis='both', which='major', labelsize=10)  # Adjust tick label size
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.savefig('mae_ctemp.png')
plt.show()
```
